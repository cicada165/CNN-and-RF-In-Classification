```python
import matplotlib.pyplot as plt
import numpy as np
  
x = [3, 2, 4, 1, 2, 4, 4]
y = [4, 2, 4, 4, 1, 3, 1]


plt.scatter(x, y, c = ['red', 'red', 'red', 'red', 'blue', 'blue', 'blue'])
x = np.linspace(0,5,100)
plt.plot(x, x, '--g')
plt.plot(x, x-0.5, '-g')
plt.plot(x, x-1, '--g')
plt.title('Observation and Maximum-Margin')
plt.show()
```
